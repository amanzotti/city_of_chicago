What happen on a station by station base?

```python
for feature in ['Air Temperature','Humidity','Barometric Pressure','Wind Speed']:
    
    sns.distplot(humidity[humidity['Station Name']== Stations[0]][feature],bins=10,label=Stations[0])
    sns.distplot(humidity[humidity['Station Name']== Stations[1]][feature],bins=10,label=Stations[1])
    sns.distplot(humidity[humidity['Station Name']== Stations[2]][feature],bins=10,label=Stations[2])
    plt.legend(loc=0)
    plt.show()
    plt.clf()

```

### Work on time stamps, making them more readable

```python
humidity['Measurement_time']= pd.to_datetime(humidity['Measurement Timestamp'])
```

```python
# dir(humidity['Measurement_time'].dt)
```

```python
humidity['Month'] = humidity['Measurement_time'].dt.month
humidity['dayweek'] = humidity['Measurement_time'].dt.dayofweek
humidity['day'] = humidity['Measurement_time'].dt.dayofyear
humidity['hour'] = humidity['Measurement_time'].dt.hour
```

```python
sns.boxplot(x="Month", y="Humidity", hue="Station Name", data=humidity);
```
